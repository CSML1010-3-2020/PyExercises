### Coordinates and Diagonals

```python
# Controlling Axes and Adding Diagonals
k = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
k = k.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
k.set(xlim=(0,100), ylim=(0,100))
for ax in k.axes.flat:
    ax.plot((0,100), (0,100), c='gray', ls='--' )
k.add_legend()
```

## Building Dashboards

A dashboard is a combination of plots.

```python
sns.set_style('darkgrid')
f, ax = plt.subplots(2,2, figsize=(15,15))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, cmap='Greens_r', ax=ax[0,0])
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, cmap='Greens_r', ax=ax[0,1])
k1.set(xlim=(-20,175))
k2.set(xlim=(-20,175))
z1 = sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year', y='CriticRating', ax=ax[1,0])
k3 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                 shade=True, shade_lowest=False, kernel='gau', \
                 cmap='Reds', ax=ax[1,1])
#ax[1,1].hist(movies.CriticRating, bins=15)
plt.show()
```

## Styling Tips

```python
sns.set_style('dark', {"axes.facecolor":"black"})
f, ax = plt.subplots(2,2, figsize=(10,10))
# Plot [0,0]
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                 shade=True, shade_lowest=True, cmap='inferno', \
                 ax=ax[0,0])
k1b = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                 cmap='cool', ax=ax[0,0])
# Plot [0,1]
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, \
                 shade=True, shade_lowest=True, cmap='inferno', \
                 ax=ax[0,1])
k2b = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, \
                 cmap='cool', ax=ax[0,1])
# Plot [1,0]
z1 = sns.violinplot(data=movies, \
                    x='Year', y='BudgetMillions', \
                    palette='YlOrRd', ax=ax[1,0])
# Plot [1,1]
k3 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                 shade=True, shade_lowest=False, cmap='Blues_r', \
                 ax=ax[1,1])
k3b = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                 cmap='gist_gray_r', ax=ax[1,1])
#ax[1,1].hist(movies.CriticRating, bins=15)
k1.set(xlim=(-20,175))
k2.set(xlim=(-20,175))
plt.show()
```

## Finishing Touches

```python
sns.set_style('whitegrid', {"axes.facecolor":"white"})
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
fig, ax = plt.subplots()
fig.set_size_inches(12, 8) # size of A4 paper
h = plt.hist(list1, bins=30, stacked=True, rwidth=10, label=mylabels)
plt.title("Movie Budget Distribution", fontsize=35, \
         color='DarkBlue')
plt.ylabel("Number of Movies", fontsize=25, color="Red")
plt.xlabel("Budget", fontsize=25, color="Green")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, \
           framealpha=1, prop={'size':20})
plt.show()
```
